## Matlab

```matlab
close all;clc;clear all;                                         
%Flauta
soundFile1 = ['/home/victor/DCO2004_LabPSC/MATERIAL/HD_03_MATLAB/Flauta.wav'];           
[vtSomFlauta, dFa1] = audioread(soundFile1);  
dFa1=dFa1/1e3;
lfft = 1000;  % Número de pontos da fft
freq1 = [0 : dFa1/lfft : dFa1/2 - dFa1/lfft];
m1fft1 = fft(vtSomFlauta,lfft)/lfft;              
m1fftuni1 = m1fft1(1:lfft/2);
figure(1)
plot(freq1,abs(m1fftuni1));
title('Flauta na frequência' )
xlabel('Frequência [kHz]')
ylabel('|G(f)|')
xlim([0 10])

%Violino                                          
soundFile2 = ['/home/victor/DCO2004_LabPSC/MATERIAL/HD_03_MATLAB/Violino.wav'];           
[vtSomVio, dFa2] = audioread(soundFile2);  
dFa2=dFa2/1e3;
lfft = 1000;  % Número de pontos da fft
freq2 = [0 : dFa2/lfft : dFa2/2 - dFa2/lfft];
m1fft2 = fft(vtSomVio,lfft)/lfft;           
m1fftuni2 = m1fft2(1:lfft/2);
figure(2)
plot(freq2,abs(m1fftuni2));
title('Violino na frequência' )
xlabel('Frequência [kHz]')
ylabel('|G(f)|')
xlim([0 15])
```

Pelo espectro do violino, notamos que o tom agudo se dá pelas fibrações rápidas, frequências maiores, já no da flauta, tom grave, as fibrações são lentas, frequências menores. Isso é notado nos gráficos, por exemplo, no gráfico do violino são vários picos, e no intervalo de frequência, praticamente não chega a zero o |G(f)|, já no gráfico da flauta, são 4 grandes picos, com frequências bem proximas.

## Python 

```matlab
soundFile1 = '/home/victor/DCO2004_LabPSC/MATERIAL/HD_03_MATLAB/Violino.wav'               
import numpy as np
import scipy.io.wavfile as wv 
import os
import matplotlib.pyplot as plt

dFa1,vtSomVio = wv.read(soundFile1)                                 
dFa1=dFa1/1e3
                          

lfft = 600 
m1fft1 = np.abs(np.fft.fft(vtSomVio,lfft))
freq1 = np.linspace(0.0,dFa1/2,lfft//2)
m1fftuni1 = m1fft1[:lfft//2]
plt.title("Violino na frequência")
plt.xlabel("Frequência (kHz)")
plt.ylabel("|G(f)|")
plt.grid()
plt.plot(freq1,m1fftuni1)
plt.xlim([0,15])
plt.show()

soundFile2 = '/home/victor/DCO2004_LabPSC/MATERIAL/HD_03_MATLAB/Flauta.wav'               

dFa2,vtSomFla = wv.read(soundFile2)                                 
dFa2=dFa2/1e3
lfft = 600 
m1fft2 = np.abs(np.fft.fft(vtSomFla,lfft))
freq2 = np.linspace(0.0,dFa2/2,lfft//2)
m1fftuni2 = m1fft2[:lfft//2]
plt.title("Flauta na frequência")
plt.xlabel("Frequência (kHz)")
plt.ylabel("|G(f)|")
plt.grid()
plt.plot(freq2,m1fftuni2)
plt.xlim([0,10])
plt.show()

```
