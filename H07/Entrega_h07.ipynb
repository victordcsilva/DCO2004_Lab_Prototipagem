```python
import numpy as np
from scipy.stats import norm
from scipy.stats import rayleigh
from scipy.stats import lognorm
from scipy.stats import rice
from scipy.stats import nakagami
from scipy.stats import weibull_max
from scipy.stats import expon

import matplotlib.pyplot as plt
#
# Parmetros da distribuio
vtMu = [0, -10, 10]                  # Valores de mdia da Gaussiana
vtVar = [1, 5, 10]                   # Valores de varincia da Gaussiana
x = np.arange(-20,20,0.1)     
#
# Variando a mdia e plotando os grficos
plt.figure(1,[15,7])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    plt.subplot(2,2,1)
    plt.plot(x, rayleigh.pdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('PDF Rayleigh')
    plt.subplot(2,2,2)
    plt.plot(x, rayleigh.cdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('CDF Rayleigh')
# Variando a varincia e plotando os grficos
mu = vtMu[0];
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, rayleigh.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('PDF Rayleigh')
    plt.subplot(2,2,4)
    plt.plot(x, rayleigh.cdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('CDF Rayleigh')
    
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()

plt.figure(2,[15,7])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    plt.subplot(2,2,1)
    plt.plot(x, norm.pdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('PDF Normal')
    plt.subplot(2,2,2)
    plt.plot(x, norm.cdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('CDF Normal')
# Variando a varincia e plotando os grficos
mu = vtMu[0];
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, norm.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('PDF Normal')
    plt.subplot(2,2,4)
    plt.plot(x, norm.cdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('CDF Normal')
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()

plt.figure(3,[15,7])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    plt.subplot(2,2,1)
    plt.plot(x, rice.pdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('PDF Rice')
    plt.subplot(2,2,2)
    plt.plot(x, rice.cdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('CDF Rice')
# Variando a varincia e plotando os grficos
mu = vtMu[0];
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, rice.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('PDF Rice')
    plt.subplot(2,2,4)
    plt.plot(x, rice.cdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('CDF Rice')
    
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()

plt.figure(4,[15,7])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    plt.subplot(2,2,1)
    plt.plot(x, expon.pdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('PDF Exponencial')
    plt.subplot(2,2,2)
    plt.plot(x, expon.cdf(x,mu), label='Mdia = {}'.format(mu))
    plt.title('CDF Exponencial')
# Variando a varincia e plotando os grficos
mu = vtMu[0];
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, expon.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('PDF Exponencial')
    plt.subplot(2,2,4)
    plt.plot(x, expon.cdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.title('CDF Exponencial')
    
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()

```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
# Distribuio
mu = 1/2                                       # Mdia

T=0.001                                      # Taxa de amostragem
x=np.arange(-5,15+T,T)                        # Eixo x       
DistExp=stats.expon.pdf(x,mu)          # Distribuio    
# Clculo da probabilidade
limite_esquerdo = np.max(np.where(x<3))
limite_direito = np.min(np.where(x>15))
indices = np.arange(limite_esquerdo+1,limite_direito)
prob1=np.sum(DistExp[indices])*T*100        # Probabilidade de um evento ocorrer no intervalo
plt.figure(1,[8,6])
plt.plot(x,DistExp,'k')                                       
plt.title('Probabilidade de = ' + str(prob1))      # Mostra valor verdadeiro de prob1
plt.fill_between(x[indices],DistExp[indices],facecolor='midnightblue')
plt.show()
```
