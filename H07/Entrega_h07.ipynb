```python

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
from sympy import *
#Matematicamente
x, f = symbols('x f')

f1=2*(exp(-(2*x)))
Po1=1-integrate(f1,(x,0,3))

# Distribuição
mu = 1/5                                       # Média
T=0.001                                      # Taxa de amostragem
x=np.arange(-3,3+T,T)                        # Eixo x       
DistExp=2*stats.expon.pdf(x,mu)          # Distribuição 
# Cálculo da probabilidade
limite_esquerdo = np.max(np.where(x<0)[0])
limite_direito = np.min(np.where(x>np.exp(-6))[0])
indices = np.arange(limite_esquerdo+1,limite_direito)
prob1=abs(np.sum(DistExp[indices])*T*100)        # Probabilidade de um evento ocorrer no intervalo

plt.figure(1,[8,6])
plt.plot(x,DistExp,'k')                                       
plt.title('Probabilidade de ' + str(Po1) )     # Mostra valor verdadeiro de prob1
plt.fill_between(x[indices],DistExp[indices],facecolor='midnightblue')
plt.show()



```
